```python
import pandas as pd
import numpy as np
import pylab as pl

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

from sklearn import neighbors, tree, naive_bayes, svm, neural_network
from sklearn import ensemble
from sklearn.svm import SVC
```

```python
dataset = pd.read_csv("mono_feat.csv")
dataset.head()
data = dataset.iloc[:, 0:827].values
normalized_data=(data-data.mean())/data.std()
target = dataset.iloc[:,827].values
```

```python
labels = ['aborigene', 'praia', 'monumento','onibus','dinossauro','elefante','flor','cavalo','montanha','comida']
labels2 = ['Aborígene', 'Praia', 'Monumento','Ônibus','Dinossauro','Elefante','Flor','Cavalo','Montanha','Comida']
```

```python
X_train, X_test, y_train, y_test = train_test_split(normalized_data, target, test_size=0.30,stratify=target)
```

```python
svm = svm.SVC(kernel = 'poly', C = 10000,gamma = 'auto')
gnb = naive_bayes.GaussianNB()
c45 = tree.DecisionTreeClassifier(criterion = 'entropy')
knn = neighbors.KNeighborsClassifier(n_neighbors = 1)
```

```python
bagging = ensemble.BaggingClassifier(base_estimator = svm)
bagging.fit(X_train, y_train)
print(bagging.score(X_test, y_test))
y_predic_bagging = bagging.predict(X_test)
dfbagging = pd.DataFrame(confusion_matrix(y_test, y_predic_bagging, labels = labels))
dfbagging.columns= labels2
dfbagging.index = labels2
dfbagging
```

```python
boosting = ensemble.AdaBoostClassifier(base_estimator = svm,algorithm='SAMME')
boosting.fit(X_train, y_train)
print(bagging.score(X_test, y_test))
y_predic_boosting = boosting.predict(X_test)
dfboosting = pd.DataFrame(confusion_matrix(y_test, y_predic_boosting, labels = labels))
dfboosting.columns= labels2
dfboosting.index = labels2
dfboosting
```

```python
randomforest = ensemble.RandomForestClassifier()
randomforest.fit(X_train, y_train)
print(randomforest.score(X_test, y_test))
y_predic_randomforest = randomforest.predict(X_test)
randomforest = pd.DataFrame(confusion_matrix(y_test, y_predic_randomforest, labels = labels))
randomforest.columns= labels2
randomforest.index = labels2
randomforest
```
