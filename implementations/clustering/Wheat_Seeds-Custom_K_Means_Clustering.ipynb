```python
import pandas as pd
from sklearn.cluster import KMeans
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
class K_Means:
    def __init__(self, k =3, tolerance = 0.0001, max_iterations = 1000000):
        self.k = k
        self.tolerance = tolerance
        self.max_iterations = max_iterations

    def fit(self, data):
        
        self.centroids = {}

        #inicializa os centróides dos agrupamentos com os k primeiros elementos (poderia ser aleatório)
        for i in range(self.k):
            self.centroids[i] = data[i]

        # começa iterações
        for i in range(self.max_iterations):
            self.classes = {}
            self.classifiedData = []
            
            for i in range(self.k):
                self.classes[i] = []

            # classifica os pontos
            for features in data:
                # calcula distância do ponto entre todos os k centróides
                distances = [np.linalg.norm(features - self.centroids[centroid]) for centroid in self.centroids]
                # classifica de acordo com a menor distância entre o ponto e o centróide
                classification = distances.index(min(distances))
                # adiciona na classificação correspondente
                self.classes[classification].append(features)
                # adiciona na classificação correspondente colocando classificação como índice final
                cl=np.array(classification+1)
                feat = np.append(features,cl).tolist()
                self.classifiedData.append(feat)

            # guarda os centróides anteriores
            previous = dict(self.centroids)

            # faz a média dos pontos das classificações para definir novos centróides
            for classification in self.classes:
                self.centroids[classification] = np.average(self.classes[classification], axis = 0)

            # verifica se os centróides não mudaram de lugar (classificações não mudaram) com base na tolerância
            isOptimal = True
            for centroid in self.centroids:
                original_centroid = previous[centroid]
                curr = self.centroids[centroid]
                if np.sum((curr - original_centroid)/original_centroid * 100.0) > self.tolerance:
                    isOptimal = False

            # para classificação se os centróides são ótimos (não mudaram)
            if isOptimal:
                break

    def pred(self, data):
        distances = [np.linalg.norm(data - self.centroids[centroid]) for centroid in self.centroids]
        classification = distances.index(min(distances))
        return classification
```
