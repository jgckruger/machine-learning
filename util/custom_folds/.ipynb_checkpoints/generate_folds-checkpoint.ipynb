```python
import pandas as pd
import math
import numpy as np
from sklearn.utils import shuffle
```

```python
# read dataset from csv file
dataset = pd.read_csv("datasets/winsconsin_breast_cancer.csv")
dataset = shuffle(dataset)
dataset.head()
```

```python
# split the dataset into custom parts
A = dataset[-int(dataset.shape[0]*.25):]
BCD = dataset[:-int(dataset.shape[0]*.25)]
B = BCD[-int(BCD.shape[0]*.33):]
CD = BCD[:-int(BCD.shape[0]*.33)]
C = CD[-int(CD.shape[0]*.50):]
D = CD[:-int(CD.shape[0]*.50)]
D
```

```python
# print the datasets to files
A.to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/A.csv", index=False)
B.to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/B.csv", index=False)
C.to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/C.csv", index=False)
D.to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/D.csv", index=False)
A.append(B.append(C)).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/ABC.csv", index=False)
A.append(B.append(D)).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/ABD.csv", index=False)
A.append(C.append(D)).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/ACD.csv", index=False)
B.append(C.append(D)).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/BCD.csv", index=False)
A.append(D).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/AD.csv", index=False)
B.append(C).to_csv("/home/kruger/4o-ano/ia/machine-learning/datasets/fold/BC.csv", index=False)

```
